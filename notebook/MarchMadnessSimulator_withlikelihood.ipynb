## Welcome to the March Madness Simulator
Uses the seedings to randomly pick a winner using a Bernoulli distribution. Allows you to create a bracket without assuming any knowledge of teams beyond relative seedings.

Tweaked the professor's code a bit. I am also considering the fact that if a higher seed team beats a lower seed team, it gets a moral boost, and their chances of winning the next game increases due to higher levels of and adrenaline.

So instead of only the prior bernoulli distribution with respect to the seed, I am also considering the likelihood, to calculate a posterior. Likelihood is actually weighted by the difference of difference in seed, only if a higher seed team beats a lower seed team.

```python
import numpy as np
from scipy.stats import bernoulli
```

```python
def winner(seeda,seedb,roundnum,wona,wonb):
    """
    Picks a winner between two teams with seeds seeda and seedb. The impact of the difference dampens in the higher rounds.
    Returns 0 for seeda and 1 for seedb
    """
    ranka = float(seeda**(1/roundnum))
    rankb = float(seedb**(1/roundnum))
    wona = float(wona**(1/roundnum))
    wonb = float(wonb**(1/roundnum))
    # higher value of wonb higher the probability of w = 1 
    p = (ranka + wonb)/(ranka + rankb + wona + wonb)
    w = bernoulli.rvs(p)
    return w
```

```python
def simulateround(seeds,roundnum,won,reg):
    """
    Simulate one round for a given set of teams (with seeds)
    """
    winners = []
    for i in range(seeds.shape[0]/2):
        w = winner(seeds[2*i],seeds[2*i+1],roundnum,won[reg,2*i],won[reg,2*i+1])
        if (w == 1):
            won[reg,2*i + w] += abs(seeds[2*i + w] - seeds[2*i])
        print str(seeds[2*i])+" vs. "+str(seeds[2*i+1])
        
        print "winner = "+str(seeds[2*i + w])
        winners.append(seeds[2*i + w])
    return np.array(winners)
```

```python
seeds = np.array([1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15])
won = np.ones((4,16), dtype='int')

print "###################"
print "Simulating Midwest"
reg = 0
print "Round 1"
mw_winners = simulateround(seeds,1,won,reg)
print "\nRound 2"
mw_winners1 = simulateround(mw_winners,2,won,reg)
print "\nRound 3"
mw_winners2 = simulateround(mw_winners1,3,won,reg)
print "\nRound 4"
mw_winners3 = simulateround(mw_winners2,4,won,reg)

print "###################"
print "Simulating West"
reg = 1
print "Round 1"
w_winners = simulateround(seeds,1,won,reg)
print "\nRound 2"
w_winners1 = simulateround(w_winners,2,won,reg)
print "\nRound 3"
w_winners2 = simulateround(w_winners1,3,won,reg)
print "\nRound 4"
w_winners3 = simulateround(w_winners2,4,won,reg)

print "###################"
print "Simulating East"
reg = 2
print "Round 1"
e_winners = simulateround(seeds,1,won,reg)
print "\nRound 2"
e_winners1 = simulateround(e_winners,2,won,reg)
print "\nRound 3"
e_winners2 = simulateround(e_winners1,3,won,reg)
print "\nRound 4"
e_winners3 = simulateround(e_winners2,4,won,reg)

print "###################"
print "Simulating South"
reg = 3
print "Round 1"
s_winners = simulateround(seeds,1,won,reg)
print "\nRound 2"
s_winners1 = simulateround(s_winners,2,won,reg)
print "\nRound 3"
s_winners2 = simulateround(s_winners1,3,won,reg)
print "\nRound 4"
s_winners3 = simulateround(s_winners2,4,won,reg)
```

```python
# FROZEN FOURS
# final four game 1 - midwest vs. west
ff1_w = winner(mw_winners3[0],w_winners3[0],5,
               won[0,np.where(seeds==mw_winners3[0])],won[1,np.where(seeds==w_winners3[0])])
if ff1_w == 0:
    print "Frozen four midwest vs. west: Midwest (Seed "+str(mw_winners3[0])+")"
    reg_ff1 = 0
else:    
    print "Frozen four midwest vs. west: West (Seed "+str(w_winners3[0])+")"
    reg_ff1 = 1
# final four game 2 - east vs. south
ff2_w = winner(e_winners3[0],s_winners3[0],5,
               won[2,np.where(seeds==e_winners3[0])],won[3,np.where(seeds==s_winners3[0])])
if ff2_w == 0:
    print "Frozen four east vs. south: East (Seed "+str(e_winners3[0])+")"
    reg_ff2 = 2
else:    
    print "Frozen four east vs. south: South (Seed "+str(s_winners3[0])+")"
    reg_ff2 = 3
```

```python
# FINALS

if ff1_w == 0:
    t1 = mw_winners3[0]
    t1_name = "Midwest"
else:
    t1 = w_winners3[0]
    t1_name = "West"
if ff2_w == 0:
    t2 = e_winners3[0]
    t2_name = "East"
else:
    t2 = s_winners3[0]
    t2_name = "South"
finals_w = winner(t1,t2,6,
                  won[reg_ff1,np.where(seeds==t1)],won[reg_ff2,np.where(seeds==t2)])
if finals_w == 0:
    print "Winner: "+t1_name+" (seed "+str(t1)+")"
else:
    print "Winner: "+t2_name+" (seed "+str(t2)+")"
```
